```python
import numpy as np
import pandas as pd

df = pd.read_csv('dota2Train.csv')

df2 = pd.read_csv('dota2Test.csv')
print(df)
print(df2)
n=2000
```

```python
#train資料的前處理


z1=np.ones(n,dtype=np.float)
z2=np.ones(n,dtype=np.float)
zz1=np.zeros((5,n),dtype=np.int)
zz2=np.zeros((5,n),dtype=np.int)
for i in range(n):
  a=0
  b=0
  for j in range(4,117):
    if df.loc[i][j] ==1:
      z1[i]=z1[i]*(j/10)
      zz1[a][i]=j
      a=a+1
    if df.loc[i][j] ==-1:
      z2[i]=z2[i]*(j/10)
      zz2[b][i]=j
      b=b+1
```

```python

z3=np.ones(n,dtype=np.float)
z4=np.ones(n,dtype=np.float)
zz3=np.zeros((5,n),dtype=np.int)
zz4=np.zeros((5,n),dtype=np.int)
for i in range(n):
  a=0
  b=0
  for j in range(4,117):
    if df2.loc[i][j] ==1:
      z3[i]=z3[i]*(j/10)
      zz3[a][i]=j
      a=a+1
    if df2.loc[i][j] ==-1:
      z4[i]=z4[i]*(j/10)
      zz4[b][i]=j
      b=b+1
```

```python
z=np.zeros(n,dtype=np.int)
for i in range(n):
  if df.loc[i][0]==-1:
    z[i]=1

train_y=pd.DataFrame({'v':z})

z=np.zeros(n,dtype=np.int)
for i in range(n):
  if df2.loc[i][0]==-1:
    z[i]=1

test_y=pd.DataFrame({'v':z})
```

```python
train_x=pd.DataFrame({
    '1T':z1,
    '-1T':z2,
    })

test_x=pd.DataFrame({
    '1T':z3,
    '-1T':z4,
    })
print(zz1)
```

```python
train_x1=pd.DataFrame({
    '0':zz1[0],
    '1':zz1[1],
    '2':zz1[2],
    '3':zz1[3],
    '4':zz1[4],
    '5':zz2[0],
    '6':zz2[1],
    '7':zz2[2],
    '8':zz2[3],
    '9':zz2[4]
    })

test_x1=pd.DataFrame({
    '0':zz3[0],
    '1':zz3[1],
    '2':zz3[2],
    '3':zz3[3],
    '4':zz3[4],
    '5':zz4[0],
    '6':zz4[1],
    '7':zz4[2],
    '8':zz4[3],
    '9':zz4[4]
    })
print(train_x1)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.neural_network import MLPClassifier
from sklearn.linear_model import LogisticRegression
```

```python
clf = MLPClassifier(hidden_layer_sizes=(8,4),learning_rate_init=0.001,random_state=1, max_iter=1000).fit(train_x1, train_y)
prey=clf.predict(test_x1)
acc = accuracy_score(test_y, prey)
print(acc)
```

```python
clf = MLPClassifier(hidden_layer_sizes=(8,4),learning_rate_init=0.001,random_state=1, max_iter=1000).fit(train_x, train_y)
prey=clf.predict(test_x)
acc = accuracy_score(test_y, prey)
print(acc)
```
