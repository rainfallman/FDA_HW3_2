這次我選擇使用dota2的資料來做

資料的話一開始最前面是1跟-1，代表是哪個隊伍獲勝，再來是遊戲類型跟遊戲是否為積分賽，有可能積分比賽的腳色又跟一般場不一樣，最後多達100多的column是被選擇的腳色
這次的data太大，因為有多達100個腳色導致資料在前處理的部分會花很多時間，導致如果我全部都train跟全都test會花費太多時間，所以我只取部分的大小來做  
然後我這次的目標是只根據雙方隊伍所選擇的腳色來判斷哪邊會輸哪邊會贏
因為一場遊戲總共10人 會有五個腳色的資料為1  
五個腳色的資料為-1 

原先我用的方式是給這5個腳色一個數字，然後我再把數字相加

後來我提升準確率的方式是將相加改成相乘，然後我還會去除10的倍數來讓數字變小
因為我覺得相加的話會比較容易出現加起來一樣數字但是每個數字有所不同的情況發生，相乘比較不會出現

我使用相乘的方式是大概提升了3%的準確率

我還有做就是不用相加也不用相乘，將10個腳色的代表數字丟進去做train，但是效果不是很好，大概輸相乘2%

然後因為這個是dota2 一般來說遊戲公司會盡量讓腳色之間平衡，所以會讓每個腳色的勝率都在50%左右，所以只用腳色代號來決定勝負的話猜測準確率也大概會在50%左右
